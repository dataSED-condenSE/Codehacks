```python
import json
import os
```

# Find contests with hacks

```python
contests = os.listdir("Contests")
contests = [file for file in contests if file.endswith(".json")]
noHacks = []
notWorking = []
hacks = []
for file in contests:
    if not file.endswith(".json"):
        continue
    if "NoHacks" in file:
        noHacks.append(file)
    else:
        hacks.append(file)
f"Contests with no hacks: {len(noHacks)}; Contests with hacks: {len(hacks)};"
```

# Get problems with hacks
For each contest with hacks, find the problems to which hacking attempts have been submitted

```python
uniqueProblems = set()
for file in hacks:
    f = open(os.path.join("Contests",file),encoding="utf-8")
    data = json.load(f)
    contestNR = file.split(".")[-2]
    for hack in data:
        letter = hack["problem"]["index"]
        uniqueProblems.add((contestNR,letter))
len(uniqueProblems)
```

Write all problems to "Problems.txt"

```python
with open('Problems2.txt', 'w') as file:
    for nr,l in uniqueProblems:
        file.write(f"{nr} {l}\n")
```
