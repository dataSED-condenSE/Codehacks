```python
import json
import os
```

# Find contests with hacks

```python
contests = os.listdir("Contests")
contests = [file for file in contests if file.endswith(".json")]
noHacks = []
notWorking = []
hacks = []
for file in contests:
    if not file.endswith(".json"):
        continue
    if "NoHacks" in file:
        noHacks.append(file)
    else:
        hacks.append(file)
f"Contests with no hacks: {len(noHacks)}; Contests with hacks: {len(hacks)};"
```

# Remove hacks with "..."

```python
uniqueProblems = set()
allHacks = []
for file in hacks:
    f = open(os.path.join("Contests",file),encoding="utf-8")
    data = json.load(f)
    contestNR = file.split(".")[-2]
    for hack in data:
        if "..." in hack["judgeProtocol"]["protocol"]:
            continue
        allHacks.append(hack)
len(allHacks)
```

```python
with open(f'codehacks.json', 'w') as fout:
    json.dump(allHacks, fout)
```
