```python
import time
import hashlib
import json
import requests
from collections import Counter, defaultdict
import time
import os
from tqdm import tqdm

# TODO: add your own details
key = "" 
secret = ""
```

## Read contests.json
read all, exclude contests that did not take place yet. Revert list such that oldest is first

```python
file = open("contests.json")
contests = json.load(file)
contests = [contest["id"] for contest in contests["result"] if contest["phase"] == "FINISHED"][::-1]
len(contests)
```

```python
file = open("contest2.json")
contests2 = json.load(file)
contests2 = [contest for contest in contests2["result"] if contest["phase"] == "FINISHED"][::-1]
len(contests2)
```

Extract contests IDs from list of all contests

```python
ids = [contest["id"] for contest in contests2 if contest["id"] not in contests]
len(ids)
```

Access the Codeforces API

```python
for contestID in tqdm(ids):
    time.sleep(5)
    t = int(time.time())
    hashString = f"123456/contest.hacks?apiKey={key}&contestId={contestID}&time={t}#{secret}".encode('UTF-8')
    h = hashlib.sha512(hashString)
    req = f"https://codeforces.com/api/contest.hacks?contestId={contestID}&apiKey={key}&time={t}&apiSig=123456{h.hexdigest()}"
    x = requests.get(req)
    try:
        result = json.loads(x.text)["result"]
    except:
        with open(f'Contests/{contestID}-NoHacks.json', 'w') as fout:
            json.dump("", fout)
        continue
    if not result:
        with open(f'Contests/{contestID}-NoHacks.json', 'w') as fout:
            json.dump("", fout)
        continue
    verdicts = []
    hacks = []
    for entry in result:
        verdicts.append(entry["verdict"])
        if entry["verdict"] == "HACK_SUCCESSFUL":
            hacks.append(entry)
    content = hacks
    with open(f'Contests/{contestID}.json', 'w') as fout:
        json.dump(content, fout)
    
```
